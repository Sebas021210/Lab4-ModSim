### Task 2

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Generador 1

```python
def generadora1(seed, n):
    mod = 2**35 - 1
    a = 5**5
    x = seed
    result = []
    for i in range(n):
        x = (a * x) % mod
        result.append(x / mod)
    return result
```

### Generador 2

```python
def generadora2(seed, n):
    mod = 2**31 - 1
    a = 7**5
    x = seed
    result = []
    for i in range(n):
        x = (a * x) % mod
        result.append(x / mod)
    return result
```

### Generador 3

```python
def generadora3(n):
    return np.random.rand(n)
```

### Creación de histogramas

```python
def histogramas(n):
    seed = 12345
    nums1 = generadora1(seed, n)
    nums2 = generadora2(seed, n)
    nums3 = generadora3(n)

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].hist(nums1, bins=np.linspace(0, 1, 11), edgecolor='black')
    axs[0].set_title(f'Generador 1 ({n} repeticiones)')
    axs[1].hist(nums2, bins=np.linspace(0, 1, 11), edgecolor='black')
    axs[1].set_title(f'Generador 2 ({n} repeticiones)')
    axs[2].hist(nums3, bins=np.linspace(0, 1, 11), edgecolor='black')
    axs[2].set_title(f'Generador 3 ({n} repeticiones)')
    plt.show()
```

```python
def histogramasAsteriscos(n):
    seed = 12345
    nums1 = generadora1(seed, n)
    nums2 = generadora2(seed, n)
    nums3 = generadora3(n)
    
    bins = np.linspace(0, 1, 11)
    hist1, _ = np.histogram(nums1, bins)
    hist2, _ = np.histogram(nums2, bins)
    hist3, _ = np.histogram(nums3, bins)
    
    def imprimir_histograma(hist, n, generador):
        print(f'Generador {generador} ({n} repeticiones):')
        for i in range(len(hist)):
            porcentaje = (hist[i] / n) * 100
            asteriscos = '*' * (hist[i] // (n // 50))
            print(f'Intervalo {bins[i]:.1f}-{bins[i+1]:.1f}: {asteriscos} ({porcentaje:.2f}%)')
        print('\n')

    imprimir_histograma(hist1, n, 1)
    imprimir_histograma(hist2, n, 2)
    imprimir_histograma(hist3, n, 3)
```

### Repeticiones

```python
# Repeticiones de 100
histogramas(100)
histogramasAsteriscos(100)
```

```python
# Repeticiones de 5000
histogramas(5000)
histogramasAsteriscos(5000)
```

```python
# Repeticiones de 100000
histogramas(100000)
histogramasAsteriscos(100000)
```
