```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

```python
# Par치metros
k = 3  # Par치metro de forma (esperando el 3er carro)
lambdas = [2, 1, 0.5]  # Diferentes valores de lambda
x = np.linspace(0, 10, 1000)  # Valores de tiempo
```

```python
# Graficar las distribuciones Gamma
plt.figure(figsize=(10, 6))

for lambda_ in lambdas:
    scale = 1 / lambda_  # El par치metro de escala es 1/lambda
    y = gamma.pdf(x, k, scale=scale)
    plt.plot(x, y, label=f'lambda={lambda_}')

plt.title('Distribuci칩n Gamma para Diferentes Valores de Lambda')
plt.xlabel('Tiempo')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
