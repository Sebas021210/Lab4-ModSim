```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetro de la distribución exponencial
lam = 1.5  # Lambda, la tasa de la distribución
n_samples = 1000  # Tamaño de la muestra
```

```python
# Método de transformación inversa para generar muestras exponenciales
def inverse_transform_exponential(lam, size):
    U = np.random.uniform(0, 1, size)  # Muestras uniformes
    X = -np.log(1 - U) / lam  # Inversa de la CDF de la distribución exponencial
    return X
```

```python
# Generar la muestra de tamaño 1000
samples = inverse_transform_exponential(lam, n_samples)

# Calcular la PDF teórica de la distribución exponencial
x_values = np.linspace(0, max(samples), 1000)
pdf_theoretical = lam * np.exp(-lam * x_values)
```

```python
# Graficar el histograma de la muestra generada y la PDF teórica
plt.figure(figsize=(10, 6))

# Histograma de la muestra generada
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b', label='Muestra generada')

# PDF teórica
plt.plot(x_values, pdf_theoretical, 'r-', lw=2, label='PDF teórica')

plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma de la muestra generada vs PDF teórica de la distribución exponencial')
plt.legend()
plt.grid(True)
plt.show()
```
