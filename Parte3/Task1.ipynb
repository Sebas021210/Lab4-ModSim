```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
p = 0.3
n_samples = 1000
```

```python
def inverse_transform_geom(p, size):
    U = np.random.uniform(0, 1, size)  # Muestras uniformes
    X = np.ceil(np.log(1 - U) / np.log(1 - p))  # Inversa de la CDF de la distribución geométrica
    return X
```

```python
samples = inverse_transform_geom(p, n_samples)

k_values = np.arange(1, max(samples)+1)
pmf_theoretical = [(1 - p)**(k-1) * p for k in k_values]
```

```python
plt.figure(figsize=(10, 6))

# Histograma de la muestra generada
plt.hist(samples, bins=np.arange(1, max(samples)+2)-0.5, density=True, alpha=0.6, color='b', label='Muestra generada')

# PMF teórica
plt.stem(k_values, pmf_theoretical, 'r', markerfmt='ro', basefmt=" ", label='PMF teórica')

plt.xlabel('k')
plt.ylabel('Probabilidad')
plt.title('Histograma de la muestra generada vs PMF teórica de la distribución geométrica')
plt.legend()
plt.grid(True)
plt.show()
```
